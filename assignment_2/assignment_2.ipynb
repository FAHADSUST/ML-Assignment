# 2) Information Gain 

Create a function that given a dataset, an attribute name and a target column name, calculates the information gain for the given attribute:

The Method should work with the following call:

Method-Call: information_gain(data=data, attribute="odor", target_attribute="poisonous")

What column has the **highest information gain**?

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy as en

# Load the data
data = pd.read_csv("data/UCI_mushroom_data.csv", sep=",")

def calculateTotalEntropy(data, attribute):
    hs = data.groupby([attribute])[attribute].count()
    print(hs)
    totalEntropy = 0
    for row in hs:
        totalEntropy += (-row/len(data))*np.log2(row/len(data))
    return totalEntropy

def calculateEntropy(df, attribute, target_attribute):
    value = df.groupby(attribute)[target_attribute].apply(lambda x: len(x)/len(df[target_attribute])*(((np.sum(x)/len(x)) * np.log2(np.sum(x)/len(x))) + ((abs(len(x)-np.sum(x))/len(x)) * np.log2(abs(len(x)-np.sum(x))/len(x)))))
    return value.sum()*-1

def information_gain(data,attribute,target_attribute): 
    hs = calculateTotalEntropy(data=data,attribute=target_attribute)
    entropy = calculateEntropy(data,attribute,target_attribute)
    return hs - entropy

def calculate_highest_gain(data,target_attribute):
    attributes = data.columns.values.tolist()
    attributes.remove(target_attribute)
    maximum = -1
    columnName = ''
    for attribute in attributes:
        gain = information_gain(data=data, attribute=attribute, target_attribute=target_attribute)
        if gain > maximum:
            maximum = gain
            columnName = attribute
    return columnName 

gain = information_gain(data=data, attribute="odor", target_attribute="poisonous")
print (gain)

print (calculate_highest_gain(data, 'poisonous'))
    
    
    
    
    
```

With methods for calculating the Information Gain, we are now ready to implement our own version of the ID3 algorithm in Python. Therefore we first need two classes that represent a single node and a single edge:

```python
class Node:
    def __init__(self):
        self.value = ""
        self.edges = []

class Edge:
    def __init__(self):
        self.label = ""
        self.node = None
```

The __Node__ class has a property for the __value__ (either an attribute or a class for the leaf nodes) and a property for the list of outgoing __edges__. This list should be filled with objects of the class __Edge__.

The __Edge__ class has a property for the __label__ of the edge as well as a property for the attached __node__ that comes at the next layer.

Since we also want to show our trained decision tree, we also give you functions to print your tree:

```python
def print_tree(root):
    print(print_tree_at_layer(root, 0))

def print_tree_at_layer(tree, layer):
    text = str(tree.value)
    text += "\n"
    if not len(tree.edges) == 0:
        for e in tree.edges:
            text += "\t" * (layer+1)
            text += e.label + ": " + print_tree_at_layer(e.node, layer + 1)
    return text
```

The function __print_tree(root)__ takes the root node (an object of class __Node__) of the tree and prints it in a typical 'console directory printing style' where each line contains one node (with a possible edge label prefix). The indentation gives an indication at which layer we currently are. __print_tree_at_layer__ is an auxiliary function.
